```python
!pip install faker
import pandas as pd
from faker import Faker
import random

# Initialize Faker
fake = Faker()

# Define the number of records
num_records = 5000

# Define the structure of the dataset
data = {
    "Policy number": [],
    "Type": [],
    "Coverage": [],
    "Coverage_Add_Ons": [],
    "Policy Status": [],
    "Validity": [],
    "Insured amount": [],
    "Premium amount": [],
    "Issuance date": [],
    "Policy holder name": [],
    "Payment status": [],
    "Payment Frequency": [],
    "Policy renewal date": [],
    "Claimant name": [],
    "Relationship": [],
    "Address": [],
    "Claimant’s occupation": [],
    "Claimant’s age": [],
    "Marital status": [],
    "Claimant’s previous claims": [],
    "Claimant’s driver history": [],
    "Emergency contact information": [],
    "Phone Number": [],
    "Email": [],
    "Registration number": [],
    "Make": [],
    "Model": [],
    "Year": [],
    "Engine/Chassis number": [],
    "Color": [],
    "Vehicle condition": [],
    "Vehicle value (market value)": [],
    "Vehicle type (sedan, SUV, etc.)": [],
    "Vehicle history (accident/repair history)": [],
    "Fuel type": [],
    "Insurance coverage type": [],
    "Odometer reading": [],
    "Vehicle safety features": [],
    "Date": [],
    "Time": [],
    "Location": [],
    "Nature of incident": [],
    "Police report": [],
    "Incident severity level": [],
    "Weather conditions at the time of the accident": [],
    "Witnesses": [],
    "Traffic condition at the time of incident": [],
    "Incident description": [],
    "Involvement of pedestrians or animals": [],
    "Cause of incident": [],
    "Damage to property outside the vehicle": [],
    "Accident scene photos/videos": [],
    "Own Damage": [],
    "Third Party": [],
    "Comprehensive": [],
    "Extent of damage": [],
    "Repair estimates": [],
    "Surveyor report": [],
    "Repair Cost": [],
    "Total Loss Evaluation": [],
    "Depreciation": [],
    "Salvage value": [],
    "Parts replaced": [],
    "Insurance settlement amount": [],
    "Repair duration": [],
    "Repair shop details": [],
    "Claim Form": [],
    "Driving License": [],
    "RC Book": [],
    "FIR": [],
    "Repair estimate / bill": [],
    "Claim amount": [],
    "Deductibles": [],
    "Previous claims": [],
    "Claim status": [],
    "Supporting documents": [],
    "Payment details": [],
    "Claim history report": [],
    "Injury details": [],
    "Medical bills": [],
    "Hospitalization details": [],
    "Treatment type": [],
    "Medical reports": [],
    "Doctor’s statement": [],
    "Prescription details": [],
    "Physiotherapy details": [],
    "Disability status": [],
    "Loss of wages": [],
    "Red flags": [],
    "Investigation": [],
    "Regulatory compliance": [],
    "Court cases": [],
    "Fraud detection": [],
    "Compliance check list": [],
    "Legal obligations (if any)": [],
    "Forensic examination (if any)": [],
    "Claims audit findings": [],
    "Claim review history": [],
    "Claims approval status": [],
    "Final settlement approval": [],
    "Internal audit report": [],
    "Claim verification": [],
    "Fraud detection approval": [],
    "Supervisor/Manager approval": [],
    "Driver’s license": [],
    "Driver’s history": [],
    "Driving experience": [],
    "Previous traffic violations": [],
    "License expiry date": [],
    "Driver’s address": [],
    "Driver’s contact details": [],
    "Driver’s medical history": [],
    "Alcohol/drug test results": [],
    "License class (if applicable)": [],
    "License Type": []
}

# Generate the data
for _ in range(num_records):
    insured_amount = fake.random_int(min=200000, max=2000000)
    premium_amount = int(insured_amount * random.uniform(0.01, 0.05))  # Premium amount as 1% to 5% of insured amount
    claim_amount = int(insured_amount * random.uniform(0.1, 0.9))  # Claim amount as 10% to 90% of insured amount
    deductibles = int(claim_amount * random.uniform(0.01, 0.1))  # Deductibles as 1% to 10% of claim amount
    vehicle_value = insured_amount * random.uniform(0.5, 1.5)  # Vehicle value related to insured amount
    repair_cost = claim_amount * random.uniform(0.5, 1.0)  # Repair cost related to claim amount
    depreciation = vehicle_value * random.uniform(0.05, 0.2)  # Depreciation related to vehicle value
    salvage_value = vehicle_value * random.uniform(0.01, 0.1)  # Salvage value related to vehicle value

    data["Policy number"].append(fake.bothify(text='POL########'))
    data["Type"].append("Car Insurance")
    data["Coverage"].append(random.choice(["Comprehensive", "Third Party", "Accidental Coverage"]))
    data["Coverage_Add_Ons"].append(random.choice(["Zero Depreciation", "Road side assistance (24*7)", "No Claim Bonus", "Consumable cover", "Engine Protection"]))
    data["Policy Status"].append(random.choice(["Effective", "Cancelled", "Suspended"]))
    data["Validity"].append(f"{fake.date_between(start_date='-1y', end_date='today')} to {fake.date_between(start_date='today', end_date='+1y')}")
    data["Insured amount"].append(insured_amount)
    data["Premium amount"].append(premium_amount)
    data["Issuance date"].append(fake.date_between(start_date='-1y', end_date='today'))
    data["Policy holder name"].append(fake.name())
    data["Payment status"].append(random.choice(["Paid", "Pending"]))
    data["Payment Frequency"].append(random.choice(["Yearly", "Monthly", "Half Yearly", "Quarterly", "Recurring"]))
    data["Policy renewal date"].append(fake.date_between(start_date='today', end_date='+2y'))
    data["Claimant name"].append(fake.name())
    data["Relationship"].append(random.choice(["Spouse", "Daughter", "Parent", "Son"]))
    data["Address"].append(fake.address())
    data["Claimant’s occupation"].append(fake.job())
    data["Claimant’s age"].append(fake.random_int(min=18, max=70))
    data["Marital status"].append(random.choice(["Married", "Single"]))
    data["Claimant’s previous claims"].append(fake.random_int(min=0, max=5))
    data["Claimant’s driver history"].append(random.choice(["Clean", "1 accident", "2 minor accidents", "One speeding ticket"]))
    data["Emergency contact information"].append(f"{fake.phone_number()}, {fake.email()}")
    data["Phone Number"].append(fake.phone_number())
    data["Email"].append(fake.email())
    data["Registration number"].append(fake.bothify(text='?? ## ?? ####'))
    data["Make"].append(random.choice(["Maruti Suzuki", "Honda", "Hyundai", "Toyota", "Ford"]))
    data["Model"].append(random.choice(["Swift", "City", "i20", "Corolla", "EcoSport"]))
    data["Year"].append(fake.random_int(min=2015, max=2022))
    data["Engine/Chassis number"].append(fake.bothify(text='??##########'))
    data["Color"].append(random.choice(["Red", "Blue", "Black", "White", "Silver"]))
    data["Vehicle condition"].append(random.choice(["New", "Used"]))
    data["Vehicle value (market value)"].append(vehicle_value)
    data["Vehicle type (sedan, SUV, etc.)"].append(random.choice(["Sedan", "SUV", "Hatchback"]))
    data["Vehicle history (accident/repair history)"].append(random.choice(["No accidents", "1 accident", "1 repair"]))
    data["Fuel type"].append(random.choice(["Petrol", "Diesel"]))
    data["Insurance coverage type"].append(random.choice(["Comprehensive", "Third Party"]))
    data["Odometer reading"].append(fake.random_int(min=10000, max=100000))
    data["Vehicle safety features"].append(random.choice(["Airbags, ABS", "Airbags, ABS, EBD", "Airbags, ABS, Reverse Sensors", "Airbags, ABS, Parking Sensors"]))
    data["Date"].append(fake.date_between(start_date='-1y', end_date='today'))
    data["Time"].append(fake.time())
    data["Location"].append(fake.city())
    data["Nature of incident"].append(random.choice(["Minor Collision", "Major Collision", "Theft", "Minor Scratch", "Accident involving pedestrian"]))
    data["Police report"].append(fake.bothify(text='FIR#####'))
    data["Incident severity level"].append(random.choice(["Low", "Moderate", "High"]))
    data["Weather conditions at the time of the accident"].append(random.choice(["Clear", "Rainy", "Cloudy", "Sunny", "Overcast"]))
    data["Witnesses"].append(f"{fake.phone_number()}, {fake.name()}")
    data["Traffic condition at the time of incident"].append(random.choice(["Heavy", "Moderate", "Light"]))
    data["Incident description"].append(random.choice(["Rear-ended car", "Hit by a truck", "Stolen vehicle", "Scratched car", "Pedestrian hit"]))
    data["Involvement of pedestrians or animals"].append(random.choice(["Yes", "No"]))
    data["Cause of incident"].append(random.choice(["Driver Error", "Speeding", "Theft", "Pedestrian crossing"]))
    data["Damage to property outside the vehicle"].append(random.choice(["Yes", "No"]))
    data["Accident scene photos/videos"].append(f"{fake.file_name(extension='jpg')}, {fake.file_name(extension='mp4')}")
    data["Own Damage"].append(fake.boolean())
    data["Third Party"].append(fake.boolean())
    data["Comprehensive"].append(fake.boolean())
    data["Extent of damage"].append(fake.random_int(min=10000, max=50000))
    data["Repair estimates"].append(repair_cost)
    data["Surveyor report"].append(fake.bothify(text='SurveyorReport###'))
    data["Repair Cost"].append(repair_cost)
    data["Total Loss Evaluation"].append(random.choice(["Total Loss", "Not a total loss"]))
    data["Depreciation"].append(depreciation)
    data["Salvage value"].append(salvage_value)
    data["Parts replaced"].append(random.choice(["Rear bumper", "Left fender", "Front left door", "Left side mirror", "Windshield", "Roof panel"]))
    data["Insurance settlement amount"].append(claim_amount - deductibles)
    data["Repair duration"].append(fake.random_int(min=1, max=10))
    data["Repair shop details"].append(f"{fake.company()}, {fake.address()}, {fake.phone_number()}")
    data["Claim Form"].append(fake.bothify(text='ClaimForm###'))
    data["Driving License"].append(fake.bothify(text='DL##########'))
    data["RC Book"].append(fake.bothify(text='RC##########'))
    data["FIR"].append(fake.bothify(text='FIR#####'))
    data["Repair estimate / bill"].append(fake.bothify(text='Estimate###'))
    data["Claim amount"].append(claim_amount)
    data["Deductibles"].append(deductibles)
    data["Previous claims"].append(fake.random_int(min=0, max=5))
    data["Claim status"].append(random.choice(["Pending", "Approved", "Rejected"]))
    data["Supporting documents"].append(f"{fake.file_name(extension='jpg')}, {fake.file_name(extension='mp4')}")
    data["Payment details"].append(fake.bothify(text='Payment###'))
    data["Claim history report"].append(fake.bothify(text='HistoryReport###'))
    data["Injury details"].append(random.choice(["Broken leg", "Head injury", "Fractured arm", "Bruises on face", "Whiplash injury"]))
    data["Medical bills"].append(fake.random_int(min=10000, max=50000))
    data["Hospitalization details"].append(f"{fake.random_int(min=1, max=10)} days in {fake.company()}")
    data["Treatment type"].append(random.choice(["Surgery and bed rest", "Surgery and observation", "Casting and Physiotherapy", "Rest and bandage", "Physiotherapy and medication"]))
    data["Medical reports"].append(fake.bothify(text='MedicalReport###'))
    data["Doctor’s statement"].append(fake.bothify(text='DoctorStatement###'))
    data["Prescription details"].append(fake.bothify(text='Prescription###'))
    data["Physiotherapy details"].append(fake.bothify(text='PhysioReport###'))
    data["Disability status"].append(fake.boolean())
    data["Loss of wages"].append(fake.random_int(min=10000, max=50000))
    data["Red flags"].append(fake.boolean())
    data["Investigation"].append(random.choice(["Under process", "Closed", "Pending"]))
    data["Regulatory compliance"].append(random.choice(["Compliant", "Non-compliant"]))
    data["Court cases"].append(fake.boolean())
    data["Fraud detection"].append(random.choice(["Passed", "Failed"]))
    data["Compliance check list"].append(random.choice(["None", "Pending"]))
    data["Legal obligations (if any)"].append(random.choice(["None", "Pending"]))
    data["Forensic examination (if any)"].append(fake.bothify(text='ForensicReport###'))
    data["Claims audit findings"].append(fake.bothify(text='AuditFindings###'))
    data["Claim review history"].append(fake.bothify(text='ReviewHistory###'))
    data["Claims approval status"].append(random.choice(["Approved", "Rejected"]))
    data["Final settlement approval"].append(fake.boolean())
    data["Internal audit report"].append(fake.bothify(text='AuditReport###'))
    data["Claim verification"].append(fake.boolean())
    data["Fraud detection approval"].append(fake.boolean())
    data["Supervisor/Manager approval"].append(fake.bothify(text='Manager###'))
    data["Driver’s license"].append(fake.bothify(text='DL##########'))
    data["Driver’s history"].append(random.choice(["Clean driving record", "One speeding ticket", "2 minor accidents", "One traffic violation"]))
    data["Driving experience"].append(fake.random_int(min=1, max=20))
    data["Previous traffic violations"].append(random.choice(["No violations", "1 speeding", "2 accidents", "1 parking violation"]))
    data["License expiry date"].append(fake.date_between(start_date='today', end_date='+5y'))
    data["Driver’s address"].append(fake.address())
    data["Driver’s contact details"].append(f"{fake.phone_number()}, {fake.email()}")
    data["Driver’s medical history"].append(random.choice(["No major medical issues", "Asthma", "Diabetes", "Hypertension"]))
    data["Alcohol/drug test results"].append(random.choice(["Passed", "Failed"]))
    data["License class (if applicable)"].append(random.choice(["LMV", "HGV"]))
    data["License Type"].append(random.choice(["Permanent", "Learning"]))

# Create a DataFrame
df = pd.DataFrame(data)

# Export the DataFrame to a CSV file
df.to_csv('insurance_dataset_corr.csv', index=False)
```
